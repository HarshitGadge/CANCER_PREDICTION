```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns




```

```python
df_cancer=pd.read_csv("cancer_data")
```

```python
x = df_cancer.drop(['target'],axis =1)
```

```python
x
```

```python
y= df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
 x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
svc_model= SVC()
```

```python
svc_model.fit(x_train,y_train)
```

```python
y_predict =svc_model.predict(x_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm ,annot=True)
```

```python
min_train =x_train.min()
```

```python
min_train =x_train.min()
```

```python
range_train =(x_train - min_train).max()
```

```python
x_train_scaled =(x_train-min_train)/range_train
```

```python
sns.scatterplot(x = x_train['mean area'], y= x_train['mean smoothness'],hue =y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y= x_train_scaled['mean smoothness'],hue =y_train)
```

```python
min_test =x_test.min()
range_test =(x_test - min_test).max()
x_test_scaled =(x_test-min_test)/range_test
```

```python
svc_model.fit(x_train_scaled,y_train)
```

```python
y_predict =svc_model.predict(x_test_scaled)
```

```python
cn = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cn,  annot = True)
```

```python
print(classification_report(y_test,y_predict))
```

```python
param_grid ={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
import pickle

def model_prediction(features, model):
    
    pickled_model = pickle.load(open(model, 'rb'))
    cancer_prediction = str((list(pickled_model.predict(features))))
    
    return str(cancer_prediction)
```

```python
model_prediction(x_test.iloc[0:1,:],'cancerprediction.pkl')
```
